# 3) Normalise the IL from the 10 ports with the calibration

Convert the python2 pickle to python3

```python
from data.sweepobj import sweepobj
#sweepobj(data_dir + 'port1').show()

datadir2 = os.getcwd() + "/opticalSwitch/py3/" 

sw = sweepobj()

import pickle
with open(data_dir + "calibration.pickle", 'rb') as f:
    loaded = pickle.load(f, encoding="latin1") 

sw.__dict__.update(loaded)
sw.save(datadir2 + 'calibration')
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
datadir2 = os.getcwd() + "/opticalSwitch/py3/" 

# Load calibration
cal = sweepobj(datadir2 + 'calibration').detector_1


# Plot the figure
f = plt.figure(figsize=(10, 4))
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax.tick_params(which='major', direction='inout', length=8, width=2, colors='k',labelsize=18)
ax.tick_params(which='minor', direction='in', length=4, width=1, colors='k',labelsize=18)

plt.plot(sweepobj(datadir2 + 'port1').wavelength * 1e9, sweepobj(datadir2 + 'port1').detector_1-cal, label="port1")
plt.plot(sweepobj(datadir2 + 'port2').wavelength * 1e9, sweepobj(datadir2 + 'port2').detector_1-cal, label="port2")
plt.plot(sweepobj(datadir2 + 'port3').wavelength * 1e9, sweepobj(datadir2 + 'port3').detector_1-cal, label="port3")
plt.plot(sweepobj(datadir2 + 'port4').wavelength * 1e9, sweepobj(datadir2 + 'port4').detector_1-cal, label="port4")
plt.plot(sweepobj(datadir2 + 'port5').wavelength * 1e9, sweepobj(datadir2 + 'port5').detector_1-cal, label="port5")
plt.plot(sweepobj(datadir2 + 'port6').wavelength * 1e9, sweepobj(datadir2 + 'port6').detector_1-cal, label="port6")
plt.plot(sweepobj(datadir2 + 'port7').wavelength * 1e9, sweepobj(datadir2 + 'port7').detector_1-cal, label="port7")
plt.plot(sweepobj(datadir2 + 'port8').wavelength * 1e9, sweepobj(datadir2 + 'port8').detector_1-cal, label="port8")
plt.plot(sweepobj(datadir2 + 'port9').wavelength * 1e9, sweepobj(datadir2 + 'port9').detector_1-cal, label="port9")
plt.plot(sweepobj(datadir2 + 'port10').wavelength * 1e9, sweepobj(datadir2 + 'port10').detector_1-cal, label="port10")
plt.xlabel("Wavelength [nm]", fontsize=18)
plt.ylabel("Transmission [dB]", fontsize=18)
plt.xlim([1530,1570])
plt.ylim([-1.5, -0.5])
plt.legend(loc='upper right')
plt.show()
f.savefig(datadir2 + "transmission.pdf", bbox_inches='tight')
```
