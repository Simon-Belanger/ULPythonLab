```python
# TUning to 1556.2 (Real)

# Set the rings reverse bias to 2V
SMU_I.source_voltage(0)
SMU_Q.source_voltage(0)
SMU_I.output_off()
SMU_Q.output_off()
# Set the heaters bias to 0
DC_I.source_voltage(0)#1.96v to align on the notch
DC_Q.source_voltage(0) ## Away from the carrier 4V
DC_I.output_on()
DC_Q.output_on()

# Set the phaseshift
PS.source_voltage(0)
```

```python
def sweep_bias(DUT, hp, bias, delay = 0):
    """
    Sweep the bais for the Device under Test (DUT)
    """
    I = []
    P = []
    for v in bias:
        time.sleep(delay)
        DUT.source_voltage(v)
        I.append(DUT.measure_current())
        P.append(hp.readPWM(2,0))
    return np.asarray(I), np.asarray(P)
```
