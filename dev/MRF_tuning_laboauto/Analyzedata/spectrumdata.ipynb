```python
import os
import numpy as np
import matplotlib.pyplot as plt


class spectrum(object):
    
    def __init__(self):
        pass
    
    def set_det1(self, filename):
        wvl,det1,det2 = np.loadtxt(filename, dtype=float)
        self.wvl = wvl
        self.det1 = det1
        
    def set_det2(self, filename):
        wvl,det1,det2 = np.loadtxt(filename, dtype=float)
        self.wvl = wvl
        self.det2 = det2
        
    def plot_align(self):
        plt.plot(self.wvl*1e9,self.det1, label='Detector1')
        plt.plot(self.wvl*1e9,self.det2, label='Detector2')
        plt.xlabel('Wavelength (nm)')
        plt.ylabel('Power (dBm)')
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.show()
        
    def norm_data(self, filename):
        wvl,det1,det2 = np.loadtxt(filename, dtype=float)
        self.datadet1 = det1 - self.det1
        self.datadet2 = det2 - self.det2
        
    def plot(self, filename=""):
        f = plt.figure()
        plt.plot(self.wvl*1e9,self.datadet1, label='Detector1')
        plt.plot(self.wvl*1e9,self.datadet2, label='Detector2')
        plt.xlabel('Wavelength (nm)')
        plt.ylabel('Power (dBm)')
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.show()
        
        if filename != "":
            np.savetxt(filename + ".txt", (self.wvl,self.datadet1,self.datadet2))
            f.savefig(filename + ".pdf")
            
        
```

```python
data_dir = os.getcwd()
s1 = spectrum()
s1.set_det1(data_dir + "\\align1\\align_1500_1600_0,02.txt")
s1.set_det2(data_dir + "\\align1\\align2_1500_1600_0,02.txt")
s1.plot_align()
s1.norm_data(data_dir + "\\V3\\fullspectrumv2_1500_1600_0,02.txt")
s1.plot(data_dir + "\\V3\\norm_fullspectrumv2_1500_1600_0,02.txt")
```
